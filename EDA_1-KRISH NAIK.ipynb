```python
df_country=pd.read_excel('Country-Code.xls')
df_country.head()
```

```python
df.columns
```

```python
final_df=pd.merge(df,df_country,on='Country Code',how='left')
```

```python
final_df.head()
```

```python
##to check data types
final_df.dtypes
```

```python
country_names=final_df.Country.value_counts().index
```

```python
country_value=final_df.Country.value_counts().values
```

```python
#pie chart - top 5 countries based on the uses of Zomato 
plt.pie(country_value[:4],labels=country_names[:4],autopct="%1.2f%%")
```

Observation: Zomato Maximum records of Transaction is from India followed by USA and United Kingdom
